```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM , Dense
from sklearn.preprocessing import MinMaxScaler

data = pd.read_excel('Book1.xlsx')

longitude_data = data['longitude'].values
latitude_data = data['latitude'].values
timestamp_data = data['timestamp'].values
speed_data = data['speed'].values

longitude_scaler = MinMaxScaler()
latitude_scaler = MinMaxScaler()
timestamp_scaler = MinMaxScaler()
speed_scaler = MinMaxScaler()

longitude_data = longitude_scaler.fit_transform(longitude_data.reshape(-1, 1))
latitude_data = latitude_scaler.fit_transform(latitude_data.reshape(-1, 1))
timestamp_data = timestamp_scaler.fit_transform(timestamp_data.reshape(-1, 1))
speed_data = speed_scaler.fit_transform(speed_data.reshape(-1, 1))

x = np.concatenate((longitude_data , latitude_data , timestamp_data , speed_data), axis=1)
x = x.reshape((-1 , 1 ,4))

train_size = int(0.8 * len(x))
train_x , test_x =  x[:train_size] , x[train_size:]
train_y,test_y = x[:train_size,:,:2],x[train_size:,:,:2]

model = Sequential()
model.add(LSTM(units=128, return_sequences=True, input_shape=(1, 4)))
model.add(LSTM(units=64))
model.add(Dense(2))
model.compile(loss='mse', optimizer='adam')

model.fit(train_x, train_y, epochs=100, batch_size=32, validation_data=(test_x, test_y))

mse = model.evaluate(test_x, test_y)
print(f'Test MSE: {mse:.2f}')

predictions = model.predict(test_x)
longitude_predictions =longitude_scaler.inverse_transform(predictions[:,0].reshape(-1,1))
latitude_predictions = latitude_scaler.inverse_transform(predictions[:,1].reshape(-1,1))
test_lon = longitude_scaler.inverse_transform(test_y[:,0].reshape(-1,1))
test_lan = latitude_scaler.inverse_transform(test_y[:,0].reshape(-1,1))

print('Longitude Predictions: ',longitude_predictions[:5])
print('Latitude Predictions: ',latitude_predictions[:5])

mean_lon=np.mean(longitude_predictions)
mean_lat=np.mean(latitude_predictions)

print(f'Mean Longitude Predictions: {mean_lon:.5f}')
print(f'Mean Latitude Predictions: {mean_lat:.5f}')
```
